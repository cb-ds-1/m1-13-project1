```python
# Using the function ratio_m_t,
# I added a new column for vectorized money/time ratio

import numpy as np

def ratio_m_t(money, time):
    '''Function with 2 arguments, money and time,
    returns ratio of money over time'''
    return (money / time)

df['money_time_ratio'] = np.vectorize(ratio_m_t)(df['money'], df['time (hr)'])
```

```python
# Using the function distance_to_helicopter,
# I added a new column for vectorized distance to helicopter coordinates

import math

def distance_to_helipad(x1, y1):
    '''Function with 2 arguments, x and y coordinates of bank,
    returns distance to helicopter at coordinates (0,0)'''
    x2, y2 = (0,0)
    return math.hypot(x2 - x1, y2 - y1)

df['distance_to_helipad'] = np.vectorize(distance_to_helipad)(df['x_coordinate'], df['y_coordinate'])
```

```python
from scipy.spatial import distance

def robbery(df, start):
    
    ''' A function to rob a maximum of banks
    and getting to 'helipad' at coordinates (0,0) within 24 hours.
    
    It takes 2 arguments, df and start. 
    
    While working on a solution, I added two columns to the original df:
    
        - money_time_ratio: calculates the 'money'/'time (hr)' ratio per bank.
        At each loop containing a sample of 100 banks, the bank to rob
        will be chosen based on the row with the largest value of this ratio.
        
        - distance_to_helipad: calculates the distance to 'helipad'.
        In the loop, this will be assigned the variable 'distance_heli'.
        
    I've excluded rows with 'money' less than $9_000.
    There are now 4297 rows × 7 columns in the updated df1
    
    For 'start', it is an integer representing a bank.
    By having a starting point at a bank, I can rob it right away.
    I chose bank 3613 because it had the max 'money'/'time (hr)' ratio.
    The start values are calculated into travel_list, t_remaining, and score.
    
    The algorithm returns 
    - total score
    - travel_list (including 'helipad'), and 
    - t_remaining 
    
    Scipy distance.euclidean is used with inputs (previous_coord, current_coord),
    where each argument is a pair of x,y coordinates,
    to calculate distance between banks.
    
    The formula v = d/t, where
    v = speed
    d = distance
    t = time
    is used to determine arithmetic calculations for distances and times.
    
    The max score output so far has been: $7_074_300, with
    travel_list[
    3613, 8562, 7074, 6104, 4499, 2331, 5933, 7560, 9546, 7343, 2037,
    613, 3803, 487, 8525, 3466, 4789, 8849, 6097, 5135, 8287, 2656,
    3005, 9378, 9653, 8436, 2346, 9583, 6528, 58, 7544, 6468, 8690,
    6740, 865, 9241, 3914, 7087, 3026, 8355, 2928, 4757, 9195, 1757,
    7258, 4696, 70, 4605, 7701, 9908, 2472, 8469, 7064, 3798, 8908,
    524, 2243, 9736, 8206, 4723, 5610, 4762, 3186, 9401, 6375, 9228,
    1447, 7583, 433, 7649, 444, 8966, 3516, 8022, 7595, 5563, 3605,
    507, 7665, 8846, 9881, 2190, 9779, 2741, 4906, 3193, 2458, 9290,
    1455, 4287, 1372, 517, 6156, 5126, 9469, 951, 6712, 1599, 9275, 557,
    7877, 8407, 7764, 5719, 8703, 5296, 1397, 6317, 6254, 279, 8231,
    5622, 7689, 1976, 2626, 6281, 5295, 6591, 9049, 6169]
    and t_remaining: 0.03000924091639881
    
    
    
    Conclusions:
    - This algorithm does reliably give scores in the mid-$600_000s.
    
    - Excluding 'money' values higher than $9_000 did not improve scores.
    
    - Changing sampling size from 100 to 200, did not improve scores,
    so I put it back at 100 to keep output time low. Experimenting with
    a sample size of 500 was reckless.
    
    I'm wondering if one could incorporate Monte Carlo to have
    more options in choosing an optimal travel_list.
    
    To make the solution work, I input the original df (not the altered df)
    and the travel_list into check_solution.
    
    With a current top score of $7_074_300, for now I'll have to settle for 
    buying an RV and travelling the world for a couple of years.
    
    '''

def robbery(df, start):
    # initial variables are set
    travel_list = []
    t_remaining = float(24)
    score = 0
    speed = 30

    # start values inputted into travel_list, t_remaining, score
    travel_list.append(start)    
    t_remaining -= df.loc[start]['time (hr)']
    score += df.loc[start]['money']

    # start bank coordinates become first previous_coord    
    previous_coord = df.loc[start]['x_coordinate'], df.loc[start]['y_coordinate']
        
    # LOOP STARTS HERE
    while t_remaining > 0.00:
        df_temp = df.sample(100)
        temp_row = df_temp.nlargest(1, 'money_time_ratio')
        bank_id = int(temp_row.iloc[0]['id'])
        distance_heli = temp_row.iloc[0]['distance_to_helipad']
    
        current_coord = temp_row.iloc[0]['x_coordinate'], temp_row.iloc[0]['y_coordinate']
        distance_bank = distance.euclidean(previous_coord, current_coord)
        
        #if not t_remaining < distance_heli/speed:
        if t_remaining >= distance_bank/ speed:
            if bank_id not in travel_list:
                #travel_list.append(bank_id)
                travel_list.append(int(temp_row.iloc[0]['id']))
                full_time = (temp_row['time (hr)'].iloc[0] + distance_bank/speed)

                t_remaining -= full_time
                score += temp_row['money'].iloc[0]

                previous_coord = current_coord
                
        else:
            travel_list.append('helipad')
            t_remaining -= distance_heli/speed
            break
            
    
    return f"Score: ${score}; travel_list: {travel_list}; and remaining time: {t_remaining} time!"


df_robbery = df
start = 3613
robbery(df_robbery, start)
```
